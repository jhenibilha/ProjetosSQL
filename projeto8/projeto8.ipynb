```python
import random

nomes = ["Ana Souza", "Carlos Lima", "João Pedro", "Mariana Alves", "Lucas Rocha", "Beatriz Mendes", "Paulo César", "Fernanda Silva", "Gustavo Dias", "Juliana Costa"]
generos = ["Masculino", "Feminino", "Outro"]
condicoes = ["Diabetes", "Hipertensão", "Asma", "Depressão", "Covid-19", "Obesidade"]
tratamentos = ["Insulina", "Beta-bloqueador", "Inalador", "Psicoterapia", "Antiviral", "Dieta"]
medicos = ["Dr. Silva", "Dra. Oliveira", "Dr. Souza", "Dra. Costa"]
hospitais = ["Hospital Central", "Clínica São João", "Santa Casa", "Hospital Vida"]
planos = ["Unimed", "Bradesco Saúde", "Amil", "SulAmérica", "Particular"]

linhas = []
for _ in range(300):
    nome = random.choice(nomes)
    genero = random.choice(generos)
    idade = random.randint(18, 90)
    condicao = random.choice(condicoes)
    tratamento = random.choice(tratamentos)
    medico = random.choice(medicos)
    hospital = random.choice(hospitais)
    plano = random.choice(planos)
    
    linha = f"('{nome}', '{genero}', {idade}, '{condicao}', '{tratamento}', '{medico}', '{hospital}', '{plano}')"
    linhas.append(linha)

blocos = [linhas[i:i+50] for i in range(0, 300, 50)]

for i, bloco in enumerate(blocos):
    print(f"-- Bloco {i+1}")
    print("INSERT INTO tratamentos (paciente_nome, genero, idade, condicao, tratamento, medico, hospital, plano_saude)\nVALUES")
    print(",\n".join(bloco) + ";\n")
```
